# Lab 1: Thao tác cơ bản với numpy, cv2
- Load ảnh: `cv2.imread()`
- Hiển thị ảnh: `cv2.imshow()` `cv2.waitKey()` `cv2.destroyAllWindows`
- Tách kênh màu: `numpy splicing`, `cv2.split()` `cv2.merge()`
- Chuyển đổi hệ màu: `cv2.cvtColor()`

```python
import numpy as np
import cv2 as cv2
```

```python
# path = r'./resources/low-contrast-1.png'
path = r'./resources/rose.jpg'

img = cv2.imread(path, cv2.IMREAD_COLOR_BGR)
```

```python
h,w,c = img.shape

blue = np.zeros_like(img)
blue[:,:,0] = img[:,:,0]

green = np.zeros_like(img)
green[:,:,1] = img[:,:,1]

red = np.zeros_like(img)
red[:,:,2] = img[:,:,2]

b,g,r = cv2.split(img)

z = np.zeros(shape=(h,w), dtype=np.uint8)
blue = cv2.merge([b,z,z])
green = cv2.merge([z,g,z])
red = cv2.merge([z,z,r])

origin = cv2.merge([b,g,r])
origin = np.stack([b,g,r], axis=2)

cv2.imshow('original', origin)
cv2.imshow('blue', blue)
cv2.imshow('green', green)
cv2.imshow('red', red)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
only_v = img_hsv[:,:,2]
```

```python
cv2.imshow('gray', img_gray)
cv2.imshow('hsv', img_hsv)
cv2.imshow('only_v', only_v)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
